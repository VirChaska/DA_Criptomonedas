```python
import pandas as pd
```

```python
bitcoin_df = pd.read_csv("Data/bitcoin.csv")
ethereum_df = pd.read_csv("Data/ethereum.csv")
tether_df = pd.read_csv("Data/tether.csv")
binancecoin_df = pd.read_csv("Data/binancecoin.csv")
ripple_df = pd.read_csv("Data/ripple.csv")
usd_coin_df = pd.read_csv("Data/USDcoin.csv")
staked_ether_df = pd.read_csv("Data/StakedEther.csv")
dogecoin_df = pd.read_csv("Data/dogecoin.csv")
cardano_df = pd.read_csv("Data/cardano.csv")
solana_df = pd.read_csv("Data/solana.csv")
```

```python
dataframes = [bitcoin_df, ethereum_df, binancecoin_df, ripple_df, dogecoin_df, cardano_df, solana_df, tether_df, usd_coin_df, staked_ether_df]
criptomonedas = ['Bitcoin', 'Ethereum', 'BinanceCoin', 'Ripple', 'Dogecoin', 'Cardano', 'Solana', 'Tether', 'USD Coin', 'Staked Ether']

dfs_combinados = []

for i, df in enumerate(dataframes):

    df['Criptomoneda'] = criptomonedas[i]
    
    df = df[['timestamp', 'Criptomoneda', 'precio', 'market_cap', 'total_volumen']]
    
    dfs_combinados.append(df)

df_combinado = pd.concat(dfs_combinados, ignore_index=True)

df_combinado = df_combinado.sort_values(by='timestamp')
df_combinado.reset_index(drop=True, inplace=True)
```

```python
df_combinado = df_combinado.rename(columns={
    'timestamp': 'fecha',
    'Criptomoneda': 'criptomoneda',
    'precio': 'precio',
    'market_cap': 'capitalizacion',
    'total_volumen': 'volumen_total'
})
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_combinado.to_csv("Data/criptomonedas.csv", index=False)
```
